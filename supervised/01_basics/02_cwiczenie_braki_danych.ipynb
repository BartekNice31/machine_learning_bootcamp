````python
import numpy as np
import pandas as pd
import sklearn
from sklearn.preprocessing import LabelEncoder

sklearn.__version__
````

````python
data = {
    'size': ['XL', 'L', 'M', np.nan, 'M', 'M'],
    'color': ['red', 'green', 'blue', 'green', 'red', 'green'],
    'gender': ['female', 'male', np.nan, 'female', 'female', 'male'],
    'price': [199.0, 89.0, np.nan, 129.0, 79.0, 89.0],
    'weight': [500, 450, 300, np.nan, 410, np.nan],
    'bought': ['yes', 'no', 'yes', 'no', 'yes', 'no']
}

df_raw = pd.DataFrame(data=data)
df_raw
````

````python
df=df_raw.copy()
````

# **Ważne- w oparciu o brak danych (uzupełnienie średnia lub mediana) wykonujemy te obliczenia na treningowych zbiorach!!!!**

````python
df
````

````python
df.isna().sum()/len(df)
````

````python
df['size'].value_counts()
````

````python
df['size'].mode()[0]
````

````python
df['size']=df['size'].fillna(df['size'].mode()[0])
df['gender']=df['gender'].fillna(df['gender'].mode()[0])
````

````python
df
````

````python
df['price']=df['price'].fillna(df['price'].median())
df['weight']=df['weight'].fillna(df['weight'].median())
````

````python
[df['price'].mean(),df['price'].median()],[df['weight'].mean(),df['weight'].median()]
````

````python
df,df.isna().sum().sum()==0
````

````python
df=df_raw.copy()
````

````python
from sklearn.impute import SimpleImputer
simple_imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
simple_imputer.fit_transform(df[['price']])
````

````python
simple_imputer.statistics_
````

````python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit_transform(df[['price','weight']])
````

````python
imputer.statistics_
````

````python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='median')
imputer.fit(df[['price','weight']])
````

````python
imputer.statistics_
````

````python
imputer=SimpleImputer(missing_values=np.nan,strategy='constant',fill_value=99.0)
imputer.fit(df[['price','weight']])
````

````python
df.select_dtypes('object')
````

````python
imputer_objects=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
imputer_objects.fit(df.select_dtypes('object').drop(columns='bought',axis=1))
````

````python
imputer_objects.statistics_
````
